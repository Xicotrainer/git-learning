```python
# Now, we'll use the zip functions with same-length iterables 
books = ["Eat that frog", "Unlimited power", "The code of Extraordinary man", "The voice of your soul"]
masters = ("Brian Tracy", "Tony Robins", "Vishem Lakiani", "Lain Garcia")

for pack in zip(masters, books):
    print(f'Master {pack[0]} wrote "{pack[1]}"')
```
